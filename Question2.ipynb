```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encode the 'Geography' column
label_encoder = LabelEncoder()
analysis_data['Geography'] = label_encoder.fit_transform(analysis_data['Geography'])

# Define features (X) and target variables (y)
X = analysis_data[['Geography', 'Price per Unit', 'Price per Volume']]
y_unit_sales = analysis_data['Unit Sales']
y_dollar_sales = analysis_data['Dollar Sales']

# Split the data into training and testing sets
X_train, X_test, y_train_unit, y_test_unit = train_test_split(X, y_unit_sales, test_size=0.2, random_state=42)
X_train, X_test, y_train_dollar, y_test_dollar = train_test_split(X, y_dollar_sales, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Train and evaluate a Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train_unit)

# Predictions and evaluation
y_pred_unit = linear_model.predict(X_test)
print("Linear Regression (Unit Sales):")
print("MSE:", mean_squared_error(y_test_unit, y_pred_unit))
print("R2 Score:", r2_score(y_test_unit, y_pred_unit))
```

```python
from sklearn.tree import DecisionTreeRegressor

# Train and evaluate a Decision Tree model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train_unit)

y_pred_dt = dt_model.predict(X_test)
print("Decision Tree (Unit Sales):")
print("MSE:", mean_squared_error(y_test_unit, y_pred_dt))
print("R2 Score:", r2_score(y_test_unit, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train and evaluate a Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train_unit)

y_pred_rf = rf_model.predict(X_test)
print("Random Forest (Unit Sales):")
print("MSE:", mean_squared_error(y_test_unit, y_pred_rf))
print("R2 Score:", r2_score(y_test_unit, y_pred_rf))
```
